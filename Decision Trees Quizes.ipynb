## Construction of Corona Decision Tree  (ID3 algorithm)
Calculating the IG for different features of Covid
<img src=https://images.upgrad.com/5db06a4e-c588-4e44-af25-3fef9c8fafc6-7.png>

```python
# Replicating the data in the image in the form of a Pandas DF
data = {1:['Fever', 'more than 14 days', 'Without Mask', 'No'],
         2:['Fever', 'more than 14 days', 'With Mask', 'No'],
         3:['Breathing Problem', 'more than 14 days', 'Without Mask', 'Yes'],
         4:['Cough', 'more than 14 days', 'Without Mask', 'Yes'],
         5:['Cough', 'less than 14 days', 'Without Mask', 'Yes'],
         6:['Cough',  'less than 14 days', 'With Mask', 'No'],
         7:['Breathing Problem', 'less than 14 days', 'With Mask', 'Yes'],
         8:['Fever', 'more than 14 days', 'Without Mask', 'No'],
         9:['Fever', 'less than 14 days', 'Without Mask', 'Yes'],
         10:['Cough', 'less than 14 days', 'Without Mask', 'Yes'],
         11:['Fever', 'less than 14 days', 'With Mask', 'Yes'],
         12:['Breathing Problem', 'more than 14 days', 'With Mask', 'Yes'],
         13:['Breathing Problem', 'less than 14 days', 'Without Mask' ,'Yes'],
         14:['Cough', 'more than 14 days', 'With Mask', 'No']
        }
data_df = pd.DataFrame(data=data.values(),
                       index = data.keys(),
                       columns=['Symptoms', 'International Travel', 
                                'Interaction with Infected Person', 
                                'Corona Test Result'
                               ]
                      )
data_df
```

```python
# First calculating the system entropy 
# Number of Yes = 9
# Number of No = 5
data_count=14
p_yes = 9/data_count
p_no = 5/data_count

system_entropy = -p_yes*np.log2(p_yes) - p_no * np.log2(p_no)
print(f'System entropy = {system_entropy:.3f}')

print('Calculating the entropy of Symptoms Feature')
# There are 3 values of feature: Fever, Cough, Breathing Problem
fever_count = 5
p_fever_yes = 2/fever_count
p_fever_no = 3/fever_count
fever_entropy = -p_fever_yes * np.log2(p_fever_yes) - p_fever_no * np.log2(p_fever_no)
print(f'Entropy of Fever = {fever_entropy:.3f}')

cough_count = 5
p_cough_yes = 3/cough_count
p_cough_no = 2/cough_count
cough_entropy = -p_cough_yes * np.log2(p_cough_yes) - p_cough_no * np.log2(p_cough_no)
print(f'Entropy of Cough = {cough_entropy:.3f}')

breathing_count = 4
p_breathing_yes = 4/breathing_count
p_breathing_no = 0/breathing_count
breathing_entropy = -p_breathing_yes * np.log2(p_breathing_yes) #- p_breathing_no * np.log2(p_breathing_no)
print(f'Entropy of Breathing Problem = {breathing_entropy:.3f}')

# Entropy of the Symptoms feature is the weighted entropy of the above
symptoms_entropy = (fever_count/data_count)*fever_entropy + \
                    (cough_count/data_count) * cough_entropy +\
                    (breathing_count/data_count) * breathing_entropy
print(f'Entropy of Symptoms Feature = {symptoms_entropy:.3f}')


print('Calculating the entropy of International Travel Feature')
# There are 2 discrete values, more_than_14days and less_than_14days
# Calculating hte entrropy of more than 14 days (greater than 14 days)
gt_count= 7
p_gt_yes = 3/gt_count
p_gt_no = 4/gt_count
gt_entropy = -p_gt_yes * np.log2(p_gt_yes) - p_gt_no * np.log2(p_gt_no)
print(f'Entropy of More than 14 days = {gt_entropy:.3f}')

#Calculating the entropy of less than 14 days
lt_count = 7
p_lt_yes = 6/lt_count
p_lt_no = 1/lt_count
lt_entropy = -p_lt_yes * np.log2(p_lt_yes) - p_lt_no * np.log2(p_lt_no)
print(f'Entropy of Less than 14 days = {lt_entropy:.3f}')

#Enttropy of the International Travel feature is the weighted average of >14days and <14days
international_travel_entropy = (gt_count/data_count)*gt_entropy + \
                                (lt_count/data_count)*lt_entropy
print(f'Entropy of the International Travel Feature = {international_travel_entropy:.3f}')

print('Calculating the entropy of Interaction with Infected Person Feature')
# There are 2 discrete values with mask and without mask
# Calculating the entropy of with mask
with_count = 6
p_with_yes= 3/with_count
p_with_no = 3/with_count
with_entropy = -p_with_yes * np.log2(p_with_yes) - p_with_no * np.log2(p_with_no)
print(f'Entropy of With Mask = {with_entropy:.3f}')

# Calculating the entropy of without mask
without_count = 8
p_without_yes = 6/without_count
p_without_no = 2 / without_count
without_entropy = -p_without_yes * np.log2(p_without_yes) - p_without_no * np.log2(p_without_no)
print(f'Entropy of without mask = {without_entropy:.3f}')

# Calculating the Weighted Average to get the entropy 
mask_entropy = (with_count/data_count) * with_entropy + (without_count/data_count)*without_entropy
print(f'Entropy of with/without mask = {mask_entropy:.3f}')

df = pd.DataFrame(data=[symptoms_entropy,
                        international_travel_entropy,
                        mask_entropy
                       ],
                  index=['Symptom', 'International_Travel', 'Mask'],
                  columns=['Entropy']
                 )
df['IG'] =  system_entropy - df['Entropy']
```

```python
df.sort_values('IG', ascending=False)
# Since Symptom has the highest IG, that would form thhe parent node, or in this case, the root node
```

```python
# Replicating the data in the image in the form of a Pandas DF
data = {1:['Fever', 'more than 14 days', 'Without Mask', 25.],
         2:['Fever', 'more than 14 days', 'With Mask', 30.],
         3:['Breathing Problem', 'more than 14 days', 'Without Mask', 46.],
         4:['Cough', 'more than 14 days', 'Without Mask', 45.],
         5:['Cough', 'less than 14 days', 'Without Mask', 52.],
         6:['Cough',  'less than 14 days', 'With Mask', 23.],
         7:['Breathing Problem', 'less than 14 days', 'With Mask', 43.],
         8:['Fever', 'more than 14 days', 'Without Mask', 35.],
         9:['Fever', 'less than 14 days', 'Without Mask', 38.],
         10:['Cough', 'less than 14 days', 'Without Mask', 46.],
         11:['Fever', 'less than 14 days', 'With Mask', 48.],
         12:['Breathing Problem', 'more than 14 days', 'With Mask', 52.],
         13:['Breathing Problem', 'less than 14 days', 'Without Mask' ,44.],
         14:['Cough', 'more than 14 days', 'With Mask', 30.]
        }
data_df = pd.DataFrame(data=data.values(),
                       index = data.keys(),
                       columns=['Symptoms', 'International Travel', 
                                'Interaction with Infected Person', 
                                'Corona Infection Percentage'
                               ]
                      )
data_df
```

```python
def std_dev(df, column):
    '''
    df: Pandas Dataframe
    column: String, the name of the column for which std dev has to be computed
    
    '''
    return np.sqrt(np.sum(np.array(df[column].mean() - df[column])**2)/len(df))
```

```python
#System Std Dev calculation

data_count = 14
#Find standard deviation of the target
#system_std = data_df['Corona Infection Percentage'].std()
system_std = std_dev(data_df,
                     'Corona Infection Percentage')
print(f'System std dev = {system_std:.3f}')

# Starting with the Symptoms Feature 
#There are 3 distinct categories
#Stdd of target for Fever
#fever_std = data_df[(data_df.Symptoms == 'Fever')]['Corona Infection Percentage'].std()
fever_std = std_dev(data_df[(data_df.Symptoms == 'Fever')],
                    'Corona Infection Percentage')
print(f'fever std ={fever_std:.3f}')

# std of Cough 
#cough_std = data_df[(data_df.Symptoms == 'Cough')]['Corona Infection Percentage'].std()
cough_std = std_dev(data_df[(data_df.Symptoms == 'Cough')],
                            'Corona Infection Percentage')
print(f'Cough std = {cough_std:.3f}')

#std of Breathing Problem
#breathing_std = data_df[(data_df.Symptoms == 'Breathing Problem')]['Corona Infection Percentage'].std()
breathing_std = std_dev(data_df[(data_df.Symptoms == 'Breathing Problem')],
                        'Corona Infection Percentage')
print(f'Breathing Problem std = {breathing_std:.3f} ')

# Overall Std of Symptoms is the weighted average of the std of individual categories
# Finding the value counts of each categories
#print("\nSymptoms category count:\n",data_df.Symptoms.value_counts())
symptom_std = (len(data_df[data_df.Symptoms=='Fever'])/data_count)*fever_std \
            + (len(data_df[data_df.Symptoms=='Cough'])/data_count)*cough_std \
            + (len(data_df[data_df.Symptoms=='Breathing Problem'])/data_count)*breathing_std

print(f'std of Symptoms feature ={symptom_std:.3f}')

print('Calculations of std of International Travel ')
# There are 2 distinct catetegories
#gt_std = data_df[(data_df['International Travel'] == 'more than 14 days')]['Corona Infection Percentage'].std()
gt_std = std_dev(data_df[(data_df['International Travel'] == 'more than 14 days')],
                 'Corona Infection Percentage')
print(f'Std of More than 14 days travel = {gt_std:.3f}')

#lt_std= data_df[(data_df['International Travel'] == 'less than 14 days')]['Corona Infection Percentage'].std()
lt_std= std_dev(data_df[(data_df['International Travel'] == 'less than 14 days')],
                'Corona Infection Percentage')
print(f'Std of Less than 14 days travel = {gt_std:.3f}')

# Overall std of International Travel is the weighted average of the std of individual categories
international_travel_std = (len(data_df[(data_df['International Travel'] == 'more than 14 days')])/data_count)*gt_std \
                        + (len(data_df[(data_df['International Travel'] == 'less than 14 days')])/data_count)*lt_std 
print(f'std of International Travel = {international_travel_std:.3f}')

print('Calculations of std of Interaction with Infected Person')
# There are 2 distinct categories
#with_mask_std = data_df[data_df['Interaction with Infected Person']=='With Mask']['Corona Infection Percentage'].std()
with_mask_std = std_dev(data_df[data_df['Interaction with Infected Person']=='With Mask'],
                        'Corona Infection Percentage')
print(f'Std of with mask = {with_mask_std:.3f}')
#without_mask_std = data_df[data_df['Interaction with Infected Person']=='Without Mask']['Corona Infection Percentage'].std()
without_mask_std = std_dev(data_df[data_df['Interaction with Infected Person']=='Without Mask'],
                           'Corona Infection Percentage')
print(f'Std of without mask = {without_mask_std:.3f}')

#Overall std of feature
interaction_std = (len(data_df[(data_df['Interaction with Infected Person']=='With Mask')])/data_count)*with_mask_std \
                + (len(data_df[(data_df['Interaction with Infected Person']=='Without Mask')])/data_count)*without_mask_std 
print(f'Std of Interaction with Infected person = {interaction_std:.3f}')

print('Calculating thhe differences in Std:')
df = pd.DataFrame(data=[system_std - symptom_std,
                        system_std - international_travel_std,
                        system_std - interaction_std
                       ],
                  columns = ['std_diff'],
                  index=['Symptoms',
                         'International Travel',
                         'Interaction with Infected Person'
                        ]
                 )
#Sorting by thhe differrences in Std
df.sort_values('std_diff',
               inplace=True,
               ascending=False
              )
df
```

Since Symptoms has the max difference in std w.r.t the system std, **the root node is selected as Symptoms**<br>

```python
# Replicating the data in the image in the form of a Pandas DF
data = {1:['Fever', 'more than 14 days', 'Without Mask', 'No'],
         2:['Fever', 'more than 14 days', 'With Mask', 'No'],
         3:['Breathing Problem', 'more than 14 days', 'Without Mask', 'Yes'],
         4:['Cough', 'more than 14 days', 'Without Mask', 'Yes'],
         5:['Cough', 'less than 14 days', 'Without Mask', 'Yes'],
         6:['Cough',  'less than 14 days', 'With Mask', 'No'],
         7:['Breathing Problem', 'less than 14 days', 'With Mask', 'Yes'],
         8:['Fever', 'more than 14 days', 'Without Mask', 'No'],
         9:['Fever', 'less than 14 days', 'Without Mask', 'Yes'],
         10:['Cough', 'less than 14 days', 'Without Mask', 'Yes'],
         11:['Fever', 'less than 14 days', 'With Mask', 'Yes'],
         12:['Breathing Problem', 'more than 14 days', 'With Mask', 'Yes'],
         13:['Breathing Problem', 'less than 14 days', 'Without Mask' ,'Yes'],
         14:['Cough', 'more than 14 days', 'With Mask', 'No']
        }
data_df = pd.DataFrame(data=data.values(),
                       index = data.keys(),
                       columns=['Symptoms', 'International Travel', 
                                'Interaction with Infected Person', 
                                'Corona Test Result'
                               ]
                      )
data_df
```

```python
# Considering Binary split as
# Branch 1 = Breathing Problem
# Branch 2 = Cough + Fever

data_count = len(data_df)
# Finding the probabilities of Target Variable in case of Breathing Problem branch
breathing_count=len(data_df[(data_df['Symptoms'] == 'Breathing Problem')])
# Probability of a Yes out of the rows hhaving Breathing Problems
p_breathing_yes = len(data_df[(data_df['Symptoms'] == 'Breathing Problem') & \
                              (data_df['Corona Test Result'] == 'Yes')])/breathing_count
# Probability of a No out of the rows hhaving Breathing Problems
p_breathing_no = len(data_df[(data_df['Symptoms'] == 'Breathing Problem') & \
                              (data_df['Corona Test Result'] == 'No')])/breathing_count


# Calculating Gini Index of Breathing Problem branch
print(f'p_breathing_yes={p_breathing_yes:.3f}, p_breathing_no={p_breathing_no:.3f}')
gini_breathing = 1 - (p_breathing_yes**2 + p_breathing_no**2)
print(f'Gini Index for Breathing ={gini_breathing:.3f}')

# Calculating the probabilities of Target Variable in case of Cough + Fever Branch
cough_fever_count=len(data_df[(data_df['Symptoms'] != 'Breathing Problem')])
# Probability of a Yes out of the rows having Cough or Fever symptom
p_cough_fever_yes = len(data_df[(data_df['Symptoms'] != 'Breathing Problem') & \
                                (data_df['Corona Test Result'] == 'Yes')])/cough_fever_count
# Probability of a No out of the rows having Cough or Fever symptom
p_cough_fever_no = len(data_df[(data_df['Symptoms'] != 'Breathing Problem') & \
                                (data_df['Corona Test Result'] == 'No')])/cough_fever_count
# Calculating the Gini index of the branch
gini_cough_fever = 1 - (p_cough_fever_yes**2 + p_cough_fever_no**2)
print(f'Gini Index for Cough + Fever = {gini_cough_fever:.3f}')

#Calculating the Gini Index as the weighted sum of the branches
gini_symptoms = (len(data_df[(data_df.Symptoms == 'Breathing Problem')])/data_count) * gini_breathing + \
                (len(data_df[(data_df.Symptoms != 'Breathing Problem')])/data_count) * gini_cough_fever
print(f'Gini Index of Symptoms = {gini_symptoms:.3f}')
```
