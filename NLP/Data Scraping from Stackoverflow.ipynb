```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
```

```python
myurl = 'https://stackoverflow.com/questions/415511/how-to-get-the-current-time-in-python'
html = urlopen(myurl).read()
html
```

```python
soupified = BeautifulSoup(html, features='html.parser' )
soupified
```

```python
question = soupified.find('div', {'id':'question-header'})
question
```

```python
question_text = question.find('a', {'class':'question-hyperlink'})
question_text
```

```python
question_text.get_text().strip()
```

```python
answer = soupified.find('div',{'class':'accepted-answer'})
answer
```

```python
answer_text = answer.find('div',{'class':'answercell'}).find('div',{'class':'s-prose'}).get_text()
answer_text
```
